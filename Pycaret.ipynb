```python
# Let is read the Dataset and import Pandas library
```

```python
import pandas as pd
flt_dly = pd.read_csv("E:\DATA scientist\R\Flight\Flight1987 csv.csv")
```

```python
# The column "FlightNum" describes the unique identification of a flight. 
# This column does not add value to the data as it is not related to departure delay. 
# Hence we drop this column
# Also let is replace the missing values
```

```python
flt_dly = flt_dly.drop(["FlightNum"],axis = 1)
flt_dly1 = flt_dly.dropna(subset = ["w","DayofMonth","DayOfWeek","DepTime","ArrTime","ActualElapsedTime","ArrDelay","DepDelay"])
```

```python
flt_dly1.isnull().sum()
```

```python
# The Federal Aviation Administration (FAA) considers a flight to be delayed when it is 15 minutes later than its scheduled time. 
# Hence in the "DepDelay" column, we will mutate the data such as , if the flight is delayed by more than 15 minutes, 
# it is considered as delayed(1), otherwise it is not delayed(0).
```

```python
import numpy as np
flt_dly1["Flight_Delayed"] = np.where(flt_dly1["DepDelay"] > 15,1,0)
```

```python
# The data wrangling process is complete.
```

```python
flt_dly1.head()
```

```python
flt_dly1 = flt_dly1.drop(["DepDelay"],axis = 1)
```

```python
# The data wrangling process is complete.
```

```python
# Now let us import all the classification algorithms from Pycaret library and set our target variable
```

```python
from pycaret.classification import *
clf = setup(data = flt_dly1, target  = "Flight_Delayed")
```

```python
# Let us compare the performance of classification algorithms 
```

```python
compare_models()
```
